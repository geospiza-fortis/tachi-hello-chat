```python
import requests
import pandas as pd

def get_search(offset, auth_token):
    headers = auth = {"authorization": auth_token}
    url = (
        "https://discord.com/api/v9/guilds/349436576037732353"
        "/messages/search"
        "?author_id=417419992531861517"
        "&content=goodnight%20chat"
        "&has=image"
        "&include_nsfw=true"
        f"&offset={offset}"
    )
    return requests.get(url, headers=headers)

res = get_search(0, auth_token).json()
```

```python
import itertools

messages = list(itertools.chain.from_iterable(res["messages"]))
pd.DataFrame(messages).head()
```

```python
messages = []
for i in range(11):
    offset = i * 25
    search = get_search(offset, auth_token)
    messages += list(itertools.chain.from_iterable(search.json()['messages']))
print(len(messages))
```

```python
df = pd.DataFrame(messages)
df
```

```python
ts_df = df[["timestamp"]].copy()
ts_df["timestamp"] = pd.to_datetime(ts_df["timestamp"])
ts_df["date"] = ts_df["timestamp"].dt.date
ts_df["hour"] = (ts_df["timestamp"].dt.hour + ts_df["timestamp"].dt.minute / 60).round(2)
ts_df.to_csv("2023-08-07-sin_goodnight_chat.csv", index=False)
```
